```python
import pandas as pd

# Step 1: Read the CSV file
df = pd.read_csv('Books_data.csv')

# Step 2: Remove the 'Wished Users' column
df = df.drop(columns=['Wished Users'])

# Step 3: Clean the 'Price (TK)' column by removing commas and converting to numeric
df['Price (TK)'] = df['Price (TK)'].replace({',': ''}, regex=True).astype(float)

# Convert 'Copies Left', 'Ratings', 'Reviews' columns to numeric, coerce errors (invalid parsing becomes NaN)
df['Copies Left'] = pd.to_numeric(df['Copies Left'], errors='coerce')
df['Ratings'] = pd.to_numeric(df['Ratings'], errors='coerce')
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Step 4: Handle missing values
# Fill missing values in numeric columns with the median of each column (this is just one option)
df['Copies Left'] = df['Copies Left'].fillna(df['Copies Left'].median())
df['Ratings'] = df['Ratings'].fillna(df['Ratings'].median())
df['Reviews'] = df['Reviews'].fillna(df['Reviews'].median())

# Check the cleaned dataframe
print(df.head())

```

```python
import pandas as pd

# Step 1: Read the CSV file
df = pd.read_csv('Books_data.csv')

# Step 2: Remove the 'Wished Users' column
df = df.drop(columns=['Wished Users'], errors='ignore')

# Step 3: Print column names to verify the existence of 'Ratings' and other columns
print(df.columns)

# Step 4: Clean the 'Price (TK)' column by removing commas and converting to numeric
df['Price (TK)'] = df['Price (TK)'].replace({',': ''}, regex=True).astype(float)

# Convert 'Copies Left', 'Ratings', 'Reviews' columns to numeric, coerce errors (invalid parsing becomes NaN)
df['Copies Left'] = pd.to_numeric(df['Copies Left'], errors='coerce')

# Check if 'Ratings' and 'Reviews' columns exist, and clean them if they do
if 'Ratings' in df.columns:
    df['Ratings'] = pd.to_numeric(df['Ratings'], errors='coerce')

if 'Reviews' in df.columns:
    df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Step 5: Handle missing values
df['Copies Left'] = df['Copies Left'].fillna(df['Copies Left'].median())

# Check the cleaned dataframe
print(df.head())
```
