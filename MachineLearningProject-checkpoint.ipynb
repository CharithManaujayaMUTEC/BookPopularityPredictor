## HANDLE MISSING VALUES

### Handling Missing Values and Data Type Conversion:

Missing values are handled in two main ways: 
Categorical columns are filled with the most frequent value (mode)
Numeric columns are filled with the median.
'Publication'missing values are replaced with the most common value using `data['Publication'].mode()[0]`. For numeric columns'Wished Users','Ratings','Reviews', missing values are filled with the median value (using `fillna(data['Wished Users'].median()`, etc.), as the median is more robust to outliers than the mean.

### Data Type Conversion:

'Wished Users','Ratings','Reviews' columns contain numeric data stored as strings, often including commas as thousand separators. To handle this:
1. First, we convert these columns to strings using `.astype(str)` to ensure that any numerical values are treated as text.
2. Then, commas are removed using `.str.replace(',', '')`, ensuring that the numbers are cleaned of formatting.
3. Finally, we convert the cleaned data to numeric using `pd.to_numeric()`. This method ensures that any non-numeric values (such as strings or malformed entries) are turned into `NaN`, which is handled by filling the missing values later with median values.

The overall goal is to ensure that all relevant columns are of the correct data type (numeric) and that missing data is appropriately imputed with meaningful statistics, maintaining the integrity and consistency of the dataset.

```python
import pandas as pd

# Load dataset
data = pd.read_csv('Books_data.csv')

# Step 1: Strip spaces from column names
data.columns = data.columns.str.strip()

# Step 2: Remove the 'Wished Users' column (since it's no longer needed)
data = data.drop(columns=['Wished Users'], errors='ignore')

# Check the result to ensure the column is removed
print(data.head())

# Step 3: Convert columns to numeric
numeric_columns = ['Price (TK)', 'Copies Left', 'Ratings', 'Reviews', 'Book Length (Pages)']
for col in numeric_columns:
    data[col] = pd.to_numeric(data[col].astype(str).str.replace(',', ''), errors='coerce')

# Step 4: Handle missing values for numerical columns by filling with the median
data['Price (TK)'] = data['Price (TK)'].fillna(data['Price (TK)'].median())
data['Copies Left'] = data['Copies Left'].fillna(data['Copies Left'].median())
data['Ratings'] = data['Ratings'].fillna(data['Ratings'].median())
data['Reviews'] = data['Reviews'].fillna(data['Reviews'].median())
data['Book Length (Pages)'] = data['Book Length (Pages)'].fillna(data['Book Length (Pages)'].median())

# Step 5: Handle missing values for categorical columns
# Fill missing values in 'Publication' column with the most frequent value (mode)
data['Publication'] = data['Publication'].fillna(data['Publication'].mode()[0])

# Verify no missing values remain in the entire dataset
print("\nAfter handling missing values:")
print(data.isnull().sum())

# Optionally, display a sample of the cleaned data
print("\nSample of cleaned data:")
print(data[['Price (TK)', 'Copies Left', 'Ratings', 'Reviews', 'Book Length (Pages)', 'Publication']].head())
```

### Outlier Detection and Handling

1. **Visualizing Outliers with Boxplots**:
   - **Purpose**: Boxplots visually highlight outliers in the dataset. Outliers appear as individual points outside the whiskers.
   - **What It Shows**: 
     - The box represents the interquartile range (IQR), covering the middle 50% of the data.
     - The whiskers extend up to 1.5 times the IQR, and points beyond these whiskers are considered potential outliers.

2. **Detecting Outliers Using the IQR Method**:
   - **Purpose**: The IQR method identifies outliers based on quartile values.
   - **Process**:
     1. **First Quartile (Q1)**: 25th percentile of the data.
     2. **Third Quartile (Q3)**: 75th percentile of the data.
     3. **IQR Calculation**: \( IQR = Q3 - Q1 \).
     4. **Outlier Bounds**:
        - **Lower Bound**: \( Q1 - 1.5 \times IQR \).
        - **Upper Bound**: \( Q3 + 1.5 \times IQR \).
     5. Data points below the lower bound or above the upper bound are classified as outliers.

3. **Removing Outliers (Optional)**:
   - **Purpose**: Outliers may be removed if they distort the dataset or negatively affect model performance.
   - **Considerations**: Outliers could indicate errors, rare events, or meaningful insights, so their removal should be carefully considered.

4. **Reviewing Data After Outlier Removal**:
   - **Purpose**: After handling outliers, reassess the dataset's integrity.
   - **Updated Statistics**: Recalculate summary statistics (mean, median, etc.) to check for any changes in data distribution. This confirms whether removing outliers improves the dataset’s quality.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Step 1: Convert all columns to numeric where applicable and handle errors
numeric_columns = ['Price (TK)', 'Copies Left', 'Wished Users', 'Ratings', 'Reviews', 'Book Length (Pages)']
for col in numeric_columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')  # Convert non-numeric values to NaN

# Step 2: Handle NaN values before outlier removal, e.g., by filling with the median or dropping them
data = data.dropna(subset=numeric_columns)  # Drop rows with NaN in any of the numeric columns

# Step 3: Detect and remove outliers using the IQR method
clean_data = data.copy()  # Keep the original data intact
for col in numeric_columns:
    Q1 = clean_data[col].quantile(0.25)
    Q3 = clean_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers
    clean_data = clean_data[(clean_data[col] >= lower_bound) & (clean_data[col] <= upper_bound)]

# Step 4: Show descriptive statistics after removing outliers
print("\nAfter removing outliers:")
print(clean_data.describe())

# Step 5: Visualizations (Replot after removing outliers)

# Histograms for all numerical columns
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(clean_data[column], kde=True, bins=30, color=sns.color_palette("Set2")[i-1])
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()

# Boxplot for all numerical columns
plt.figure(figsize=(10, 5))
sns.boxplot(data=clean_data[numeric_columns])
plt.title('Boxplot for Numerical Columns (After Removing Outliers)')
plt.show()

# Scatter plot (pairplot) for all numerical columns
sns.pairplot(clean_data[numeric_columns].dropna(), diag_kind='kde')
plt.suptitle('Pairplot for Numerical Columns (After Removing Outliers)', y=1.02)
plt.show()
```

## Correlation Analysis:

In this step, we analyze the relationships between numerical variables such as Price (TK), Copies Left, Wished Users, Ratings, and Reviews. The correlation matrix helps us understand the degree to which these variables are related to each other.

Positive Correlation: Indicates that as one variable increases, the other tends to increase as well.
Negative Correlation: Indicates that as one variable increases, the other tends to decrease.
No Correlation: A value close to zero implies that the variables are not strongly related.
We visualize the correlation matrix using a heatmap for easier interpretation. The color intensity represents the strength of the correlation, and the annotations show the exact correlation values

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Handle missing values
# You can drop rows with NaN values in the columns used for correlation computation
data_clean = data[['Price (TK)', 'Copies Left', 'Wished Users', 'Ratings', 'Reviews']].dropna()

# Step 2: Calculate the Correlation Matrix for the cleaned data
correlation_matrix = data_clean.corr()

# Step 3: Display the Correlation Matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Step 4: Visualize the Correlation Matrix using a Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt='.2f', cbar=True)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```
